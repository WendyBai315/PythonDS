1. Python function pass arguments by assignment. 
2. PYthon function assign reference other than values to local variable. 

```python
class testInt():
    def __init__(self,num):
        self.n = num 
    
    def add_one(self, x):
        x += 1
        print("object id in the function :{}".format(id(x)))
```

```python
o = testInt(10)
print ("object id before sending to the function id: {} ".format(id(o.n)))
o.add_one(o.n)
print (o.n)  # should be 10
```

That means when you call a function add_one(x), it will first create a local variable name x. Once you modified the value of local x, it will point to a new PyObject with value = 11.

## Solutions

If you want to make global changes to a immutable variable, like in the first integar example:

    1. using global variables. Not recommended but works
    2. return and re-assign the value back <best practise>
        * you can return a tuple of values
        * assignment expressions: new in Python3.8 that allows you to use returned function call values in 
            different conditions.
    3. Using Objects' attributes.
    4. Using mutable objects: list & dictionary

```python
class testList():
    def __init__(self,fruits):
        self.fruits = fruits 
    def add_one(self, fruits, new_fruit):
        fruits.append(new_fruit)
        print("object id in the function :{}".format(id(fruits)))
```

```python
fruit_backet = ["apple", "banana", "orange"]
of = testList(fruit_backet)
print ("object id before sending to the function id: {} ".format(id(of.fruits)))
print("before adding fruit: {}".format(of.fruits))
of.add_one(of.fruits, "pineapple")
print (of.fruits)  # should be ["apple", "banana", "orange", "pineapple" ]
```

Below demonstrate in the first example, how to implement with object's attributes 

```python
class testInt():
    def __init__(self,num):
        self.n = num 
    
    def add_one(self):
        self.n += 1
        print("object id in the function :{}".format(id(self.n)))
```

```python
testobj = testInt(10)
print(id(testobj.n))
```

```python
testobj.add_one()
print (testobj.n)
```

It still point to a different object as int is immutable. However, the new object assigned back to the object's variable n
